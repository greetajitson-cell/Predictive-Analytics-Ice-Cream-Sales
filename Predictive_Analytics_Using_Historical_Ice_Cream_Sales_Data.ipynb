```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('ice_cream.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.iloc[:,1])
plt.title("Monthly Ice Cream Sales Trend")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.show()
```

```python
X = np.arange(len(df)).reshape(-1,1)
y = df.iloc[:,1].values
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
future_months = np.arange(len(df), len(df)+12).reshape(-1,1)

future_sales = model.predict(future_months)

print(future_sales)
```

```python
plt.figure(figsize=(10,5))

plt.plot(X,y,label='Historical Sales')

plt.plot(future_months,future_sales,
         label='Forecasted Sales')

plt.legend()
plt.show()
```

```python
score = model.score(X,y)

print("R2 Score:", score)
```

A predictive analytics model was developed using historical ice cream sales data. Linear Regression was used to forecast future sales trends. The model successfully identified patterns in the data and generated future sales predictions
